```python
import numpy as np
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer() 
from nltk import edit_distance
import re 
import pandas as pd 
import nltk
import matplotlib.pyplot as plt 
import re 
from bs4 import BeautifulSoup
import requests 
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
url='https://www.politico.com/story/2017/01/full-text-donald-trump-inauguration-speech-transcript-233907'
```

```python
# import all the stop words 
from nltk import word_tokenize
from nltk.corpus import stopwords
swords=stopwords.words('english')
print(swords)
from nltk.corpus import stopwords
stoplist = set(stopwords.words("english")) 
```

```python
plt.style.use('ggplot')
def ulr_check_in(url):
    x=[]
    y=[]
    """Function that takes in any url and returns the
    frequency distribution of the words"""
    header = {'User-Agent':'Chrome/86.0.4240.198 Safari/537.36'} 
    r = requests.get(url,headers=header)
    soup=BeautifulSoup(r.content,'lxml')
    stripped_text=soup.get_text()
    text = re.sub(r'[\r|\n|\r\n]',' ',stripped_text)
    
    tok = nltk.word_tokenize(text)
    tok_no_sword = [wd for wd in tok if wd not in stoplist]
    tok_no_sword = [wd.lower() for wd in tok_no_sword]
    tok_no_sword = [wd for wd in tok_no_sword if wd not in stoplist if wd.isalpha()]
    
    fdist = nltk.FreqDist(wd for wd in tok_no_sword)
    
    for word,count in fdist.most_common(5):
        x.append(word)
        y.append(count)
    plt.figure(figsize=(10,8))
    sns.barplot(x=y,y=x,color='#3C3B6E')  
    plt.title('President Trump Innagural Speech Top Word Used',size=18)
    plt.xlabel('Word Frequency')
    plt.ylabel('Top Words')
        
    
    lex_div=len(set(tok_no_sword))/len(tok_no_sword)
    print('Lexical  Diversity {}'.format(lex_div))
    
    
    clean_sent = ' '.join(tok_no_sword)
    wordCld = WordCloud(background_color='white').generate(clean_sent)
    plt.figure(figsize=(8,9))
    plt.imshow(wordCld)
    plt.axis('off')
    plt.title('President Trump innagural Speech Word Cloud',size=16)
    plt.show()
    
    #print(lex_div2)
    # Low lex div - Not a sign of good writer or good speaker
    # They dont vary words very much, uses same words over and over because 
    # They dont have good vocabulary. Hence doesnt hold the interst of reader 
    # Higher lexican diferesity holds the interest of the reader better
    
    
    return 
```

```python
ulr_check_in(url) # Trump 
```

```python
url_bd='https://www.politico.com/news/2021/01/20/joe-biden-inauguration-speech-transcript-full-text-460813'
```

```python
ulr_check_in(url_bd) # Biden
```
